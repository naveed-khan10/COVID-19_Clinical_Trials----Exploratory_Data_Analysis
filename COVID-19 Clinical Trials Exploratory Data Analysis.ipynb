**Solve following question by using conditional statements**


** How many studies have an enrollment greater than a certain threshold?**

```python
threshold = 170

(df["Enrollment"] > threshold).value_counts()
```

```python
great_th_thresh = df["Enrollment"] > threshold
great_th_thresh
```

```python
df_great_thre = df[great_th_thresh]
df_great_thre
```

```python
count_studies_above_threshold = df_great_thre.shape[0]
count_studies_above_threshold
```

```python
print("Number of studies with enrollment greater than", threshold, ":", count_studies_above_threshold)
```

**How many clinical trials started after January 1, 2020?**

```python
df["Start Date"]
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Start Date']
```

```python
df[df['Start Date'] > pd.Timestamp('2020-01-01')].shape
```

```python
# Convert 'Start Date' column to datetime format, handling different date formats
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Filter trials started after January 1, 2020
count_trials_after_2020 = df[df['Start Date'] > pd.Timestamp('2020-01-01')].shape[0]

print("Number of clinical_trials after_ January 1, 2020:", count_trials_after_2020)
```
